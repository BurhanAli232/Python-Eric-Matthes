- remove the extra space from string like 'pyhton' and ' python' these are two strings but most of people think this is one string

## 1-lstrip()
- remaove extra space from left side



```python
name='  Burhan'
print(name.lstrip())
```

## 2-rstrip()
- remove extra space from right side


```python
name='Burhan '
print(name.rstrip())
```

## 3-strip()
- remove the extra space from both side

```python
name=' Burhan '
print(name.strip())
```

```python
name=input("Please enter your name with space: ")
print(f'welcome {name.strip()}')
```
