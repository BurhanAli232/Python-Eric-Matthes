# Strings

## A *String* is simply a series of characters enclosed within single quotes(' ') or double quotes(" ").
### Strings can include letters, numbers, symbols, and spaces.
### But a small difference between single and double quotes is that single quotes are often used for shorter strings or when the string contains double quotes, while double quotes are used for longer strings or when the string contains single quotes.
#### Examples of strings:

### Single | Double quotes String

```python
single_quote_string = 'Hello, Anmol!' # Using single quotes
double_quote_string = "Python is fun." # Using double quotes
print(single_quote_string)
print(double_quote_string)

```

### Third quotes string

```python
string_difference = 'He said, "Python is awesome!"' # Using double quotes inside single quotes
print(string_difference)
string_difference="Hey what's up?"# Using single quotes inside double quotes
print(string_difference)
# You can also use triple quotes (''' ''' or """ """) for multi-line strings or docstrings.
multi_line_string = '''This is a string that spans
welcome Anmol the world of python programming
multiple lines.'''
print(multi_line_string)
```

### Concatination

```python
# Now we see the concatination of string
first_name=input('Please Enter your first name ')
second_name=input('Please Enter the second number ')
full_name=first_name+' '+second_name
print(full_name)
print(f'Welcome',full_name.title())
print(f'Welcome',full_name.upper())
```

### String Meathod

```python
# Some changing case in a string with meathods
name='burhaN'
print(name.title()) # Title Case
print(name.upper()) # Upper Case
print(name.lower()) # Lower Case
```
