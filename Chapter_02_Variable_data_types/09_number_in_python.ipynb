### Integer

- You can add(+), subtract(-), multiply(*), and divide(/) integers in python.

```python
# for add simple we use '+' sign also enter value which we want to add
2+3
```

```python
# for subtraction we use '-' sign also enter value which we want to subtract
3-1
```

```python
# For multiplication we use '*' sign also enter value which we want to multiply
4*2
```

```python
# For division we use '/' sign also enter value which we want to divide
9/4
```

```python
# for power we use '**' sign also enter value which we want to power
2**3
```

```python
# for multiple operations we use () to group the operations
(2+3)*4
```

### Floats

- Python calls any number with a decimal point a float.

```python

# Python calls any number with a decimal point a float.
print(0.1 + 0.2)
print(3.0 - 1.5)
print(2.5 * 4.0)
print(5.0 / 2.0)

# To perform floor division and obtain an integer result, you can use the // operator.
print(5 // 2)
print(4 // 2)
print(5.0 // 2.0)
print(5.0 // 2)
```

# Avoiding type error with the str() Function 

```python
age = 23
message = "Happy " + age + "rd Birthday!"
print(message)
```
