## 2.4-Name Case:
- Store a person's name is a variable and then print that person's name in lowercase,upercase and titlecase

```python
name=input("Enter your name: ")
print(f'Your name is in lowercase is: {name.lower()}')
print(f'Your name is in uppercase is: {name.upper()}')
print(f'Your name is in titlecase is: {name.title()}')
```

## 2.7-Stripping Name:
- Store a person's name and include some whitespace character combination, "\t" and "\n" at least once.
- print the name once, so the whitespce around the name is displayed then print the name using each of the three stripping function, lstrip(), rstrip() and strip().

```python
nmae='   Burhan\tAli    '
print(f'Name with whitespace:{nmae}')
print(f'Name after lstrip():{nmae.lstrip()}')
print(f'Name after rstrip():{nmae.rstrip()}')   
print(f'Name after strip():{nmae.strip()}')
```
